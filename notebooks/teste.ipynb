```python
import pandas as pd
```

```python
colunas_analise = ['TOTAL_VISUALIZACOES', 'TOTAL_LIKES', 'TOTAL_COMENTARIOS']
```

```python
for coluna in colunas_analise:
    tipos = {
        'ID_VIDEO': 'string',
        'TURNO_EXTRACAO': 'string',
        'INDICE_TURNO_EXTRACAO': 'string',
        coluna: 'float64'
    }
    colunas_df = ['data_extracao', 'ID_VIDEO', 'TURNO_EXTRACAO', 'INDICE_TURNO_EXTRACAO',  coluna]
    base = pd.read_csv(
        '../data/ouro_csv/total_visualizacoes_por_semana/assunto_cities_skylines/total_visualizacoes_por_semana.csv', 
        sep='|',
        parse_dates=['data_extracao'],
        usecols=colunas_df,
        dtype=tipos
    )
    base = base.query('ID_VIDEO == "EuUoqqXA6jI" and data_extracao >= "2023-10-19" and data_extracao <= "2023-10-27"  ')
    base = base.sort_values(by=['data_extracao', 'INDICE_TURNO_EXTRACAO'])
    base[f'{coluna}_DESLOCADO'] = round(base[coluna].shift(1), 0)
    base.fillna('0', axis=1, inplace=True)
    base[['ID_VIDEO', 'TURNO_EXTRACAO', ]] = base[['ID_VIDEO', 'TURNO_EXTRACAO']].astype('string')
    base[[coluna , 'INDICE_TURNO_EXTRACAO', f'{coluna}_DESLOCADO']] = base[[coluna, 'INDICE_TURNO_EXTRACAO', f'{coluna}_DESLOCADO']].astype('int32')
    base[['data_extracao']] = base[['data_extracao']].astype('datetime64[ns]')
    base[f'{coluna}_VARIACAO'] =  base[coluna] - base[f'{coluna}_DESLOCADO']
    print(coluna)
    base = base[['data_extracao', 'INDICE_TURNO_EXTRACAO', 'TURNO_EXTRACAO', 'ID_VIDEO', coluna , f'{coluna}_DESLOCADO', f'{coluna}_VARIACAO']]
    display(base)
    print()
        
```
