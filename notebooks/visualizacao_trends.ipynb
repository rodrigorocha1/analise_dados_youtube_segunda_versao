# Métricas Trends

- Categorias Populares por data de extração

```python
base_categoria = base_trends[['DATA_EXTRACAO', 'ID_CATEGORIA','TURNO_EXTRACAO', 'INDICE_TURNO_EXTRACAO', 'TOTAL_VISUALIZACOES']].query('DATA_EXTRACAO =="2023-10-16" ').sort_values(by=['ID_CATEGORIA', 'INDICE_TURNO_EXTRACAO'])
base_categoria 
```

```python
base_categoria.groupby('ID_CATEGORIA').agg(
    TOTAL_VISUALIZACOES=('TOTAL_VISUALIZACOES', 'max')
).reset_index().sort_values(by=['TOTAL_VISUALIZACOES'], ascending=False)
```

```python
base_max = base_categoria.groupby([ 'ID_CATEGORIA']).agg(
    TOTAL_VISUALIZACOES=('TOTAL_VISUALIZACOES', 'max')
).reset_index()
```
